## **Step 1: Load the Dataset and Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
```

```python
df = pd.read_csv('/content/apple_quality.csv')
df.head()
```

## Step 2: Exploratory Data Analysis (EDA)
Exploring data, preprocessing, and visualization

```python
# A_ID's are useless, so we can drop them
df.drop(columns = ['A_id'], inplace=True)
df.head()
```

```python
# Let's fix the acidity column
df['Acidity'].tail()
```

```python
# Dropping the text entry and changing datatype of the column
df = df.drop(4000)
df['Acidity'] = df['Acidity'].astype(np.float64)
```

```python
# Describe data
df.describe()
```

```python
sns.histplot(df['Size'])
```

```python
# Label encoding (good = 1 and bad = 0)
le = LabelEncoder()
df['Quality']= le.fit_transform(df['Quality'])
```

## Step 3: Split the data for our model

```python
# Split data into dependent/independent variables
x = df.iloc[:, :-1].values # Everything except Quality
y = df.iloc[:, -1].values  # Quality

# Split data into test/train set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = True)
```

## Step 4: Train models

```python
clf1 = KNeighborsClassifier(n_neighbors=3)
clf1.fit(X_train, y_train)

clf2 = svm.SVC(random_state=0)
clf2.fit(X_train, y_train)
```

## Step 5: Predict and evaluate the models

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred1))
print(metrics.accuracy_score(y_test, y_pred2))
```
